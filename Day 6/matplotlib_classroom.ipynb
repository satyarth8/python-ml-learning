```python
# linear PLOT
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[9,8,7,6,5]
plt.plot(x,y)
plt.show()
```

```python
#bar plot
plt.bar(x,y,color='r',width=0.2,align="edge",edgecolor='b')
plt.xlabel("Satyarth")
plt.ylabel("Prakash")
plt.title("Ploting a bar ",fontsize=32)
plt.plot()
```

```python
#excercise

x=["python","c","c++","java"]
y=[90,70,60,82]
z=[10,20,30,40]
plt.xlabel("Languages",color='r',fontsize=32)
plt.ylabel("Speed (in ms) ",color='r',fontsize=32)
plt.bar(x,y,color='r',edgecolor="b")
plt.bar(x,z,color='g',edgecolor="b")
plt.plot()
```

```python
import matplotlib.pyplot as plt

x= ['A', 'B', 'C', 'D']
y=[10,20,10,30]
plt.pie(y,labels=x)
plt.title('Pie Chart')
plt.show()
```
