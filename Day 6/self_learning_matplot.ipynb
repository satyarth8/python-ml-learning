```python
import matplotlib.pyplot as plt

# data to plot
categories = ['Apples', 'Oranges', 'Bananas', 'Grapes']
values = [5, 3, 6, 4]

# create bar chart
plt.bar(categories, values)

# add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('My Bar Chart')

# show the chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# data to plot
x = [1, 2, 3, 4, 5]
y = [3, 1, 4, 2, 5]

# create stem plot
plt.stem(x, y)

# add labels and title
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('My Stem Plot')

# show the plot
plt.show()
```

```python
import numpy as np

# Generate some random data
data = np.random.randint(0, 100, 50)

# Extract the stems and leaves
stems = np.floor_divide(data, 10)
leaves = np.mod(data, 10)

# Create the stem-and-leaf plot
for i in np.unique(stems):
    print(f"{i}: {', '.join(map(str, leaves[stems == i]))}")
```
